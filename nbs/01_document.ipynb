```python
from dataclasses import dataclass, field
from typing import Literal
```

```python
Language = Literal["en", "fr"]
```

```python
@dataclass
class Contact:
    email: str = ""
    phone: str = ""
    linkedin: str = ""
    github: str = ""

    def as_dict(self) -> dict:
        return {
            "email": self.email,
            "phone": self.phone,
            "linkedin": self.linkedin,
            "github": self.github,
        }
```

```python
@dataclass
class SkillCategory:
    category: str = ""
    items: list[str] = field(default_factory=list)

    def as_dict(self) -> dict:
        return {
            "category": self.category,
            "items": self.items,
        }
```

```python
@dataclass
class Experience:
    role: str = ""
    company: str = ""
    location: str = ""
    start_date: str = ""
    end_date: str = ""
    bullets: list[str] = field(default_factory=list)

    def as_dict(self) -> dict:
        return {
            "role": self.role,
            "company": self.company,
            "location": self.location,
            "start_date": self.start_date,
            "end_date": self.end_date,
            "bullets": self.bullets,
        }
```

```python
@dataclass
class Project:
    name: str = ""
    description: str = ""
    technologies: list[str] = field(default_factory=list)
    link: str = ""

    def as_dict(self) -> dict:
        return {
            "name": self.name,
            "description": self.description,
            "technologies": self.technologies,
            "link": self.link,
        }
```

```python
@dataclass
class Education:
    degree: str = ""
    institution: str = ""
    location: str = ""
    start_date: str = ""
    end_date: str = ""
    details: list[str] = field(default_factory=list)

    def as_dict(self) -> dict:
        return {
            "degree": self.degree,
            "institution": self.institution,
            "location": self.location,
            "start_date": self.start_date,
            "end_date": self.end_date,
            "details": self.details,
        }
```

```python
@dataclass
class Certification:
    name: str = ""
    issuer: str = ""
    date: str = ""
    link: str = ""

    def as_dict(self) -> dict:
        return {
            "name": self.name,
            "issuer": self.issuer,
            "date": self.date,
            "link": self.link,
        }
```

```python
from pathlib import Path

@dataclass
class CVDocument:
    language: Language = "en"

    name: str = ""
    title: str = ""
    location: str = ""
    summary: str = ""


    photo_path: str = f"file://{Path("static/photo.jpeg").resolve()}"
    linkedin_logo_path: str = f"file://{Path("static/linkedin-svgrepo-com.svg").resolve()}"
    github_logo_path: str = f"file://{Path("static/logo-github-svgrepo-com.svg").resolve()}"

    contact: Contact = field(default_factory=Contact)
    skills: list[SkillCategory] = field(default_factory=list)
    experience: list[Experience] = field(default_factory=list)
    projects: list[Project] = field(default_factory=list)

    education: list[Education] = field(default_factory=list)
    certifications: list[Certification] = field(default_factory=list)

    def is_french(self) -> bool:
        return self.language == "fr"

    def is_english(self) -> bool:
        return self.language == "en"

    def set_language(self, language: Language) -> None:
        self.language = language

    def set_photo(self, photo_path: str) -> None:
        self.photo_path = photo_path

    def assets(self) -> dict:
        return {
            "linkedin_logo": self.linkedin_logo_path,
            "github_logo": self.github_logo_path,
        }

    def as_dict(self) -> dict:
        return {
            "language": self.language,
            "name": self.name,
            "title": self.title,
            "location": self.location,
            "summary": self.summary,
            "contact": self.contact.as_dict(),
            "photo": self.photo_path,
            "assets": self.assets(),
            "skills": [skill.as_dict() for skill in self.skills],
            "experience": [exp.as_dict() for exp in self.experience],
            "projects": [proj.as_dict() for proj in self.projects],
            "education": [edu.as_dict() for edu in self.education],
            "certifications": [cert.as_dict() for cert in self.certifications],
        }
```

```python
def cv_template(cv: CVDocument) -> dict:
    d = cv.as_dict()

    return {
        "Informations personnelles": {
            "Nom": d["name"],
            "Titre": d["title"],
            "Localisation": d["location"],
            "Photo": d["photo"],
        },

        "Résumé": d["summary"],

        "Contact": {
            "Email": d["contact"]["email"],
            "Téléphone": d["contact"]["phone"],
            "LinkedIn": d["contact"]["linkedin"],
            "GitHub": d["contact"]["github"],
        },

        "Compétences": {
            skill["category"]: skill["items"]
            for skill in d["skills"]
        },

        "Expérience": {
            f"{exp['role']} @ {exp['company']}": {
                "Durée": (
                    f"{exp['start_date']} - {exp['end_date']}"
                    if exp["start_date"] and exp["end_date"]
                    else exp["start_date"] or exp["end_date"]
                ),
                "Localisation": exp["location"],
                "Technologies": "",
                "Détail": exp["bullets"],
            }
            for exp in d["experience"]
        },

        "Projets": {
            proj["name"]: {
                "Technologies": ", ".join(proj["technologies"]),
                "Détail": [
                    item for item in [proj["description"], proj["link"]]
                    if item
                ],
            }
            for proj in d["projects"]
        },

        # Important: template uses "Formation", not "Éducation"
        "Formation": {
            edu["degree"]: {
                "Institution": edu["institution"],
                "Localisation": edu["location"],
                "Année": (
                    f"{edu['start_date']} - {edu['end_date']}"
                    if edu["start_date"] and edu["end_date"]
                    else edu["start_date"] or edu["end_date"]
                ),
                "Détail": edu["details"],
            }
            for edu in d["education"]
        },

        "Certifications": {
            cert["name"]: {
                "Émetteur": cert["issuer"],
                "Année": cert["date"],
            }
            for cert in d["certifications"]
        },

        "assets": {
            "linkedin_icon": d["assets"]["linkedin_logo"],
            "github_icon": d["assets"]["github_logo"],
        },
    }
```

```python
#| export
from jinja2 import Environment, FileSystemLoader
from weasyprint import HTML, CSS
```

```python
def fix_photo_path(data: dict) -> dict:
    photo = data.get("Informations personnelles", {}).get("Photo")

    if photo:
        abs_path = Path(photo).resolve()
        data["Informations personnelles"]["Photo"] = f"file://{abs_path}"

    return data
fix_photo_path(cv)
print(cv["Informations personnelles"]["Photo"])
```

```python
#| export
def render_cv_html(
    cv: CVDocument,
    template_name: str = "cv_fr.html",
    theme_css: str = "cv2_fr.css",
    templates_dir: str | Path = "../templates",
) -> str:
    templates_dir = Path(templates_dir)

    env = Environment(
        loader=FileSystemLoader(templates_dir),
        autoescape=True,
    )

    template = env.get_template(template_name)

    return template.render(
        data=cv_template(cv),
        theme_css=theme_css
    )
```

```python
#| export
def html_to_pdf(
    html: str,
    output_path: str | Path = "../output/cv.pdf",
    css_path: str | Path = "../templates/cv_fr.css",
) -> Path:
    output_path = Path(output_path)
    css_path = Path(css_path)

    output_path.parent.mkdir(parents=True, exist_ok=True)

    HTML(string=html).write_pdf(
        target=str(output_path),
        stylesheets=[CSS(filename=str(css_path))],
    )

    return output_path
```
