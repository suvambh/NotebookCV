```python
from weasyprint import HTML
from pathlib import Path
from datetime import datetime


def generate_lom_pdf(lom_text: str, company: str, role: str, output_dir=Path("../outputs/lom/")):

    Path(output_dir).mkdir(exist_ok=True)

    safe_company = company.replace(" ", "_")
    safe_role = role.replace(" ", "_")

    filename = f"{safe_company}_{safe_role}_Lom_suvam.pdf"
    output_path = Path(output_dir) / filename

    today = datetime.now().strftime("%d %B %Y")
    lom_text = lom_text.replace("{company}", company)\
                         .replace("{role}", role)
    html_content = f"""
    <html>
    <head>
        <meta charset="utf-8">
        <style>
            body {{
                font-family: "Georgia", serif;
                margin: 80px;
                line-height: 1.7;
                font-size: 8pt;
                color: #222;
            }}

            .header {{
                text-align: right;
                margin-bottom: 40px;
            }}

            .subject {{
                margin-top: 20px;
                margin-bottom: 30px;
                font-weight: bold;
            }}

            .salutation {{
                margin-bottom: 20px;
            }}

            .paragraph {{
                margin-bottom: 18px;
                text-align: justify;
            }}

            .closing {{
                margin-top: 40px;
            }}

            .signature {{
                margin-top: 60px;
            }}
        </style>
    </head>
    <body>

        <div class="header">
            Paris, le {today}
        </div>

        <div class="subject">
            Objet : Candidature au poste de {role}
        </div>

        <div class="salutation">
            Madame, Monsieur,
        </div>

        <div class="paragraph">
            {lom_text.replace("\n\n", "</div><div class='paragraph'>")}
        </div>

        <div class="signature">
            Suvam Bhattacharya
        </div>

    </body>
    </html>
    """

    HTML(string=html_content).write_pdf(str(output_path))

    return str(output_path)
```
